## Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import re
import twint
from textblob import TextBlob
from datetime import datetime
import nest_asyncio
import nltk
from nltk.corpus import stopwords
```

### Question¶
Extract data from twitter on the 3 presidential candidates (use their names to search for them), Remove any data that isnt talking about them. Perform a sentiment analysis on them, let us see who has the most polarity (positive tweets). To do this, calculate the average for each individual.

```python
# Names of the presidential candidates for 2023 Election
candidate_name = {"Bola Ahmed Tinubu": "Bola Tinubu OR \ BAT",
                 "Peter Obi": "Obidient",
                 "Atiku Abubakar": "Atikulate"}
```

```python
# Configure twint
def twintConfig(search_string):
    c = twint.Config()
    c.Search = search_string[1]
    c.Limit = 1500
    c.Lang = "en"
    c.Pandas = True
    c.Pandas_au = True
    c.Pandas_clean =True
    c.Since = "2020-01-01"
    c.Until = "2023-01-18"
    c.Hide_output =True
    twint.run.Search(c)
```

```python
# defining a function to run twint and save result to dataframe
def Run_Twint(search_vals):
    # set empty dataframe for join
    out_df = pd.DataFrame()
    
    for candidate in search_vals.items():
        print("running for search item:"+ candidate[0]+"\n")
        print("Search string: "+ candidate[1]+ "\n")
        
        # run twint
        twintConfig(candidate)
        
        #get DataFrame
        tweets_df = twint.storage.panda.Tweets_df
        
        # join Dataframes and create "candidate" column
        tweets_df["candidate"] = candidate[0]
        out_df = pd.concat([out_df, tweets_df])
        
    return out_df 
```

```python
# Apply the nest_asyncio to avoid python looping error
nest_asyncio.apply()
tweets_df = Run_Twint(candidate_name)
```

tweets_df["language"].unique()

```python
tweets_df = tweets_df[tweets_df["language"].isin(["en"])]
```

```python
def clean_text(text):  
    pat1 = r'@[^ ]+'                   #@signs
    pat2 = r'https?://[A-Za-z0-9./]+'  #links
    pat3 = r'\'s'                      #floating s's
    pat4 = r'\#\w+'                     # hashtags
    pat5 = r'&amp '
    pat6 = r'[^A-Za-z\s]'         #remove non-alphabet
    combined_pat = r'|'.join((pat1, pat2,pat3,pat4,pat5, pat6))
    text = re.sub(combined_pat,"",text).lower()
    return text.strip()
```

```python
tweets_df["tweet"] = tweets_df["tweet"].apply(clean_text)
```

```python
# Dropping duplicated tweets but keep the first occurence
tweets_df = tweets_df.drop_duplicates(subset=["tweet", "candidate"], keep="first")
```

```python
tweets_df.head()
```

```python
tweets_df.shape
```

```python
# Download stopwords from the Natural Language toolkit
nltk.download('stopwords')
```

```python
# Removing stop words from DataFrame
stop = stopwords.words("english")
tweets_df["tweet_without_stopwords"] = tweets_df["tweet"].apply(lambda x: " ".join([word for word in x.split() if word not in (stop)]))
```

```python
tweets_df.head()
```

```python
print("Running sentiment process")
# creating two new columns(polarity and subjectivity)
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

tweets_df["Subjectivity"] = tweets_df["tweet_without_stopwords"].apply(getSubjectivity)
tweets_df["Polarity"] = tweets_df["tweet_without_stopwords"].apply(getPolarity)
```

```python
tweets_df[["Subjectivity", "tweet_without_stopwords"]]
```

```python
# Creating a column to show if the tweet is positive, negative or neutral
def analysis(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"
tweets_df["Analysis"] = tweets_df["Polarity"].apply(analysis)
```

## Question 1: who has the most polarity (positive tweets). To do this, calculate the average for each individual.

```python
tweets_df.groupby("candidate")["Polarity"].mean().plot(kind="bar", color = "pink");
plt.title("Presidential Candidate with Most Polarity")
plt.savefig('Presidential Candidate with Most Polarity.png')
plt.show();
```

## Question 2: the candidate with the most subjective tweets (Emotional tweets). To do this, calculate the average for each individual.

```python
tweets_df.groupby("candidate")["Subjectivity"].mean().plot(kind="bar", color = "pink");
plt.title("Presidentila Candidate with Most Subjectivity")
plt.savefig('Presidential Candidate with Most Subjectivity.png')
plt.show();
```
